### Airflow in the clouds

Using a managed solution allows you to push many of these responsibilities to a vendor, allowing you to focus on actually building your Airflow DAGs rather than on building and maintaining the required infrastructure. However, managed solutions may not always be flexible enough for your needs, if you have complicated requirements.

## Airflow on Azure

### Deploying Airflow in Azure

#### Picking services

* Webservers and scheduler: Azure Container Instances (ACI) or Azure Kubernetes Service (AKS). For webserver, we can also sue Azure App Service.
* Airflow metastore: Azure SQL Database.
* DAGs: Azure File Storage, as Files Storage volumes can be mounted directly into the containers running App Service and ACI.
* Data and Logs storage: Azure Blob or Data Lake Storage

#### Designing the network

We want to expose the Airflow webserver to the internet so we can access it remotely. Other components must stay in a private network.
