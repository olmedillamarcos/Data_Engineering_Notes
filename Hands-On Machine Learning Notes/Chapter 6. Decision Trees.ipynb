Decision Trees are versatile ML algorithms that, like SVM, can be used for both classification and regression. It is the fundamental component of Random Forests.

It can do categorical and numerical classification. 

For the most part, decision trees are pretty intuitive to work with: you start at the top and work your way down to you get to a point where you can't go any further.


### Going from a raw table of data to a decision tree



We start by checkin how well each feature predicts the target variable.



Because none of the leaf nodes are 100% "YES Heart Disease" or 100% "NO HEART Disease", they are all considered <b>impure</b>

To determine which separatio is best, we need to measure and compare <b>impurity</b>

### Gini Impurity




Then, take the weighted average of the gini impurity for each feature.

1) Calculate all of the Gini impurity Scores

2) If the node itself has the lowest score, then there is no point in separating the patients any more and it becomes a leaf node.

3) IF separating the data result in an improvement, then pick the separation with the lowest impurity value.

#### What if we had numerical data instead of categorical?



1) Sort the patients by weight, lowest to highest

2) Calculate the average weight for all adjacent patients



3) Calculate the impurity values for each average weight

If the feature we are comparing is ranked, we calculate the impurity for each different rank.
