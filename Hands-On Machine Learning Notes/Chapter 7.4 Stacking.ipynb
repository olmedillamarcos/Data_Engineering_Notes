# Stacking

To train the blender, we use a hold-out set: the training set is split into two subsets &rarr; one is used to train the predictors in the first layer

<div>

</div>

Then, the layer predictors are used onto the second subset (held-out). For this example, we are using three predictorsm which will predict three values: these three values for each instance are then used as a new training set, keeping the target values.

The blender is trained on the new training set, learning to predict the target value given the first layer's prediction.
