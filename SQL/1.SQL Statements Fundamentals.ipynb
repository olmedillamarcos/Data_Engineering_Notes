## SELECT DISTINCT

- Sometimes a table contains a column that has duplicate values, and you may find yourself in a situation where you only want to list the unique/distinc values.

- The DISTINCT keyword can be used to return only the distinct values in a column -> unique values inside a column

SELECT DISTINCT (column) FROM table

## COUNT

- Return the number of input rows that match a specific condition of a query

SELECT COUNT (*) FROM table

## ORDER BY

- You can use ORDER BY to sort rows based on a column value, in either ascending or descending order.

SELECT columns

FROM TABLE

ORDER BY column ASC/DESC


- Notice ORDER BY towards the end of a query, since we want to do any selection and filtering first, before finally sorting. (if left blank, it will use ASC by default)

- You can also ORDER BY multiple columns (useful for duplicate entries)

SELECT column1,column2,column3 FROM table

ORDER BY column1,column3; --> order by column1 first, then column2

----

SELECT column1,column2,column3 FROM table

ORDER BY column1 DESC,column3 ASC;

## LIMIT

- The LIMIT command allows us to limite the number of rows retrned for a query.


SELECT column1,column2

FROM table

ORDER BY column2 DESC

LIMIT nrows


## BETWEEN

- the BETWEEN operator can be used to match a value against a range of values: value BETWEEN low AND high

- same as 
value >= low AND value <= high

- You can also combine BETWEEN with the NOT logical operator:

value NOT BETWEEN low AND high

- BETWEEN can also be used with dates, Note that you need to format dates in the ISO 8601 standar format, YYYY-MM-DD

date BETWEEN '2007-01-01' AND '2007-02-01'

## IN

* In certain cases you want to check for multiple possible value options, for instance, if a user's name shows up IN  a list of known names.
- We can use the IN  opperator to create a condition that checks to see if a value is included in a list of multiple options

value IN (option1,option2)

SELECT color FROM table

WHERE color IN ('red','blue')


## LIKE and ILIKE

- What happens if we want to match against a general pattern in a string?

Ex: all emaills ending in '@gmail.com', all names that begin with an 'A'

* The LIKE operator allows us to perform pattern matching against string data with the use of wildcard characters:

* Percent %: Matches any sequence of characters
* Underscore \_: Matches any single character


- All names that begin with an 'A':

WHERE name LIKE 'A%'

- All names that end with an 'a':

WHERE NAME LIKE '%a'

* LIKE is case sensitive, while ILIKE is not



* Get all Mission Impossible films

WHERE title LIKE 'Mission Impossible \_'

* Imagine string codes in the format 'Versoin#A4', etc...

WHERE value LIKE 'Version#_ _'
