# Timestamps and Extract

PostgreSQL can hold date and time information:

   * TIME - contains only time
    
   * DATE - contains only date
    
   * TIMESTAMP - Contains date and time
    
   * TIMESTAMPTZ - Contains date, time and timezone
    

Some operators:

* TIMEZONE - shows timezone you're working with: SHOW TIMEZONE

* NOW() - shows timestamp with time zone: SELECT NOW() 

* TIMEOFDAY() - shows timestamp as a string (month, year, etc) - SELECT TIMEOFDAY() 

* CURRENT_TIME / CURRENT DATE - shows current time/date: SELECT CURRENT_

Let's explore extracting information from a time based data type using:

* EXTRACT()

* AGE()

* TO_CHAR()

#### EXTRACT()

Allows you to 'extract' or obtain a sub-component of a date value:

    - YEAR
    - MONTH
    - DAY
    - WEEK
    - QUARTER
    
EXTRACT(YEAR FROM date_col)


#### AGE()

Calculates and return the current age given a timestamp:

AGE(date_col) 

Returns 13 years 1 mon 5 days 01:34:13.003423

#### TO_CHAR()

General function to convert data types to text

* Useful for timestamp formatting
   
    * TO_CHAR(datecol,'mm-dd-yyyy')
    * [Data Type Formatting Functions](https://www.postgresql.org/docs/12/functions-formatting.html)

# SubQuery

A sub query allows you to construct complex queries, essentially performing a query on the results of another query.

The syntax is straightforward and involves two SELECT statements




SELECT student,grade

FROM test_scores

WHERE grade > (SELECT AVG(grade) 
                FROM test_scores)
                
The subquery if performed first since it is inside the parenthesis.

We can also use the IN operator in conjunction with a subquery to check against multiple results returned:


Ex: subquery that operates on a separate table

SELECT student,grade

FROM test_scores

WHERE student IN
(SELECT student FROM honor_roll_table)

#### EXISTS

The EXISTS operator is used to test for existence of rows in a subquery. Typically a subquery is passed in the EXISTS() function to check if any rows are returned with the subquery

SELECT column_name

FROM table_name

WHERE EXISTS

(SELECT column_name FROM table_name WHERE condition)

# Self-Join

A self-join is a query in which a table is joined to itself. It is useful for comparing values in a colum of rows within the same table.

SELECT tableA.col,tableB.col

FROM table AS tableA

JOIN table AS tableB ON tableA.some_col = tableB.other_col

---




SELECT emp.name, report.name

FROM employees AS emp

JOIN employees AS report ON emp.emp_id = report.report_id
